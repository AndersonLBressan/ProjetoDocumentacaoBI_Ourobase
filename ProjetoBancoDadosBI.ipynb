```python
# Conexão com SQL Server

import pandas as pd
import pyodbc
from sqlalchemy import create_engine, text


```

```python
def conecta_sql_server():
    server = 'DESKTOP-F0DI39E'
    database = 'Ourobase_CadastroDocumentacaoBI' 
    
    conn_str = (
        f"mssql+pyodbc://{server}/{database}"
        "?driver=ODBC+Driver+17+for+SQL+Server"
    ); 
    
    engine = create_engine(conn_str)
    return engine
```

### Script de criação das tabelas (DDL em Python)



```python
def criar_tabelas(engine):
    with engine.begin() as conn:
        # dim_Clientes
        conn.execute(text("""
        IF OBJECT_ID('dbo.Clientes', 'U') IS NOT NULL DROP TABLE dbo.Clientes;
        
        CREATE TABLE dbo.Clientes(
          int_codCliente INT IDENTITY (1,1) PRIMARY KEY,
          str_Cliente VARCHAR(150) NOT NULL,  
          bit_StatusAtivo BIT,
          str_DescrInativo VARCHAR(255)   
        );                  
                          
      """))
        
```

```python
if __name__ == "__main__":
    engine = conecta_sql_server()
    criar_tabelas(engine)
    print("Tabelas criadas com sucesso!")
```

#### Função para limpar as tabelas (TRUNCATE)

```python
def limpar_tabelas(engine):
    with engine.begin() as conn:
        conn.execute(text("TRUNCATE TABLE Fato_ProjetoBI_Atendimento"))
        conn.execute(text("TRUNCATE TABLE Dim_ProjetoBI"))
        conn.execute(text("TRUNCATE TABLE Dim_Servidor"))
        conn.execute(text("TRUNCATE TABLE Dim_Usuario"))
        conn.execute(text("TRUNCATE TABLE dbo_Clientes"))

```

#### Carga das dimensões via Excel

```python
caminho_do_arquivo = f'dbo_Clientes.xlsx'
def carga_dbo_Clientes(engine, caminho_do_arquivo):
    df = pd.read_excel(caminho_do_arquivo)
    df.to_sql('dbo_Clientes', con=engine, if_exists='append', index=False)
```

#### Fluxo principal (ETL completo)

```python
if __name__ == "__main__":
    engine = conecta_sql_server()

    limpar_tabelas(engine)

    carga_dbo_Clientes(engine, 'dbo_Clientes.xlsx')
    # carga_dim_usuario(engine, 'Dim_Usuario.xlsx')
    # carga_dim_servidor(engine, 'Dim_Servidor.xlsx')
    # carga_dim_projeto(engine, 'Dim_ProjetoBI.xlsx')
    # carga_fato(engine, 'Fato_ProjetoBI_Atendimento.xlsx')

    print("Carga concluída com sucesso!")

```
